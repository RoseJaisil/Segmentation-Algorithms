Import SegAutoMaskPredictor 

```python
from metaseg import SegAutoMaskPredictor
```

Segmenting using vit_1, vit_h and vit_b

```python
results = SegAutoMaskPredictor().image_predict(
    source = 'Picture1.png',
    model_type = 'vit_b', # vit_l, vit_h, vit_b
    points_per_side=16,
    points_per_batch=64,
    min_area=0,
    output_path ="output.png",
    show = False,
    save = True,

)
```
